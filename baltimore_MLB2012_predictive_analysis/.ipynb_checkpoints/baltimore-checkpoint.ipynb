### Converted the unordered day of week and month into ordered date

```r
baltimore$ordered_day_of_week <- with(data = baltimore,
                                      ifelse ((day_of_week == 'Monday'),1,
                                              ifelse ((day_of_week == 'Tuesday'),2,
                                                      ifelse((day_of_week == 'Wednesday'),3,
                                                             ifelse((day_of_week == 'Thursday'),4,
                                                                    ifelse((day_of_week == 'Friday'),5,
                                                                           ifelse((day_of_week == 'Saturday'),6,7)))))))


```

```r
baltimore$ordered_day_of_week <- factor(baltimore$ordered_day_of_week, level=1:7,
                                       labels=c('Mon','Tue','Wed','Thur','Fri','Sat','Sun'))
```

```r
table(baltimore$ordered_day_of_week)
```

```r
baltimore$ordered_month <- with(data = baltimore,
                               ifelse((month == 'APR'),4,
                                     ifelse((month == 'MAY'),5,
                                           ifelse((month == 'JUN'),6,
                                                 ifelse((month == 'JUL'),7,
                                                       ifelse((month == 'AUG'),8,9))))))
```

```r
baltimore$ordered_month <- factor(baltimore$ordered_month, levels=4:9,
                                 labels=c('Apr','May','Jun','Jul','Aug','Sep'))
```

### Created a united promotion column
Baltimore promoted their game by giving free shirts, caps or bobbleheads, but those promotions was rarely happned so that it is better to put 3 promotions as one promotion.

```r
baltimore$promotion_all <- with(data=baltimore,
                               ifelse((cap == 'YES'),1,
                                     ifelse((shirt == 'YES'),1,
                                           ifelse((bobblehead == 'YES'),1,0))))
```

```r
baltimore$promotion_all <- factor(baltimore$promotion_all, levels=0:1,
                                 labels=c('NO','YES'))
```

### Devided data into Test set and Training set in order to predict analysis

```r
set.seed(1234)
training_test <- c(rep(1, length = trunc((2/3)*nrow(baltimore))),
                   rep(2, length = (nrow(baltimore) - trunc((2/3)*nrow(baltimore)))))
baltimore$training_test <- sample(training_test)
baltimore$training_test <- factor(baltimore$training_test, 
                                   levels=c(1,2), 
                                   labels=c("Train","Test"))
```

```r
baltimore.train <- subset(baltimore, training_test == "Train")
print(str(baltimore.train))
baltimore.test <- subset(baltimore, training_test == "Test")
print(str(baltimore.test))
```

```r
my.model <- {attend ~ ordered_month + ordered_day_of_week + promotion_all}
train.model.fit <- lm(my.model, data = baltimore.train)
print(summary(train.model.fit))
baltimore.train$predict_attend <- predict(train.model.fit)
baltimore.test$predict_attend <- predict(train.model.fit, newdata = baltimore.test)
```

```r
baltimore.plotting.frame <- rbind(baltimore.train, baltimore.test)
```

```r
group.labels <- c("No_promotion","Promotion")
group.symbols <- c(21,24)
group.colors <- c("black","black")
group.fill <- c("black","red")
xyplot(predict_attend/1000 ~ attend/1000 | training_test,
      data = baltimore.plotting.frame, groups = promotion_all, cex = 2,
      pch = group.symbols, col = group.colors, fill = group.fill,
      layout = c(2,1), xlim=c(40,45), ylim = c(40,45),
      aspect = 1, type = c("p","q"),
      panel = function(x,y, ...)
          {panel.xyplot(x,y,...)
          panel.segments(40,40,45,45,col="black",cex=2)},
      strip = function(...) strip.default(..., style=1),
       xlab = "Acutal Attendance(Thousands)",
       ylab = "Predicted Attendance(Thousands)",
      key = list(space="top",
                text = list(rev(group.labels),col = rev(group.colors)),
                col = rev(group.colors),
                fill = rev(group.fill)))
```

```r
my.model.fit <-lm(my.model, data = baltimore)
print(summary(my.model.fit))
```

```r
print(anova(my.model.fit))
cat("\n","Estimated Effect of Promotion on Attendance: ",
   round(my.model.fit$coefficients[length(my.model.fit$coefficients)],
        digits = 0), "\n", sep="")
```
